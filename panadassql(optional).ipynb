```python
# Dependencies and Setup
# SQL Alchemy
from sqlalchemy import create_engine
# Pandas
import pandas as pd
# Matplotlib
import matplotlib.pyplot as plt
# NumPy
import numpy as np
```

```python
# Create Engine 
#database_path 
engine = create_engine("postgresql://postgres:postgresql@localhost:5432/sqlhomeworkbadou")
conn = engine.connect()
```

```python
# Query All Records in the the departments Table
departments_data = pd.read_sql("SELECT * FROM departments", conn)
departments_data
```

```python
# Query All Records in the the dept_emp Table
dept_emp_data = pd.read_sql("SELECT * FROM dept_emp", conn)
dept_emp_data.head()
```

```python
# Query All Records in the the employees Table
employees_data = pd.read_sql("SELECT * FROM employees", conn)
employees_data.head()
```

```python
# Query All Records in the the salaries Table
salaries_data = pd.read_sql("SELECT * FROM salaries", conn)
salaries_data.head()
```

```python
# Query All Records in the the titles Table
titles_data = pd.read_sql("SELECT * FROM titles", conn)
titles_data.head()
```

```python
# let combined the data
combined_data=pd.merge(salaries_data, titles_data, on ="emp_no", how="inner")
combined_data.head()
```

```python
group_data=combined_data.groupby(['title']).mean()
group_data.head()
```
